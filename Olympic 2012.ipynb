```python
import pandas as pd
df=pd.read_excel('Olympic 2012 Medal Tally.xlsx')
df
```

```python
# removing unwanted columns
df.drop(['Unnamed: 0'],inplace=True,axis=1)
```

```python
df
```

```python
# renaming the necessary columns
df.rename(columns={'Unnamed: 1':'Country','Unnamed: 2':'Country Code','Unnamed: 3':'Year','Unnamed: 4':'Gold','Unnamed: 5':'Silver','Unnamed: 6':'Bronze'},inplace=True)
print(df)
```

```python
# removing unnecessary rows
df.drop([df.index[0],df.index[1]],inplace=True)
df
```

```python
# countries with more than 20 gold medals 
df1=df[df['Gold']>20]
print(df1['Country'])
```

```python
# medal tally
df[['Gold','Silver','Bronze']]
```

```python
# countries with the corresponding number of gold medals
df[["Country",'Gold']]
```

```python
df2=df[df['Gold']==df['Gold'].max()]
print(df2)
```

```python
# country with maximum number of gold medals 
df2['Country']
```

```python
df['Total Medals']=df['Gold']+df['Silver']+df['Bronze']
df
```

```python
# countries with corresponding total number of medals they achieved
df[['Country','Total Medals']]
```
